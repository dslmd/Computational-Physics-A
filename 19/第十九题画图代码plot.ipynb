```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

f = open('wavefunc_E=-9.85.txt')
s = f.read()
p = list(float(i) for i in s.split())

p1 = p

f = open('wavefunc_E=-9.87.txt')
s = f.read()
p = list(float(i) for i in s.split())

p2 = p

f = open('wavefunc_E=-9.85.txt')
s = f.read()
p = list(float(i) for i in s.split())

p3 = p

x = np.arange(-5000,5500)/10000


plt.rcParams['figure.figsize'] = (20.0, 6.0)
# 画第1个图：折线图
plt.subplot(131)
plt.plot(x,p1,label="E=-9.85")
plt.xlabel("x")
plt.ylabel("$\phi$")
plt.title("E=-9.85")

# 画第2个图：散点图
plt.subplot(132)
plt.plot(x,p2,label="E=-9.87")
plt.xlabel("x")
plt.title("E=-9.87")

# 画第3个图：饼图
plt.subplot(133)
plt.plot(x,p3,label="E=-9.90")
plt.xlabel("x")
plt.title("E=-9.90")
```
