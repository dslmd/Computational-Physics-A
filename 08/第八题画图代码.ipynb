```python
import matplotlib.pyplot as plt
import numpy as np
import math
plt.figure(num=2,dpi=100)
x = np.linspace(0,2,500)

y1 = np.sqrt(x+np.sqrt(x))

plt.plot(x,y1,color = 'g',label = '$\sqrt{x+\sqrt{x}}$')         #label每个plot指定一个字符串标签

plt.legend(loc=' best') 


plt.show()
```

```python
from scipy.integrate import quad
import numpy as np

def integrand(x):
    return np.sqrt(x+np.sqrt(x))

Result = quad(integrand, 0, 2)

print(Result[0])
```

```python
from scipy import integrate

def f(x,y,z,u,v):
    return 6-x**2-y**2-z**2-u**2-v**2

Result = integrate.nquad(f, [[0, 0.9],[0, 0.8],[0,0.9],[0,2],[0,1.3]])

print(Result[0])
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math
plt.figure(num=2,dpi=100)
x = np.linspace(100,10**8,1000000)
y = 1/np.sqrt(x)


plt.axes(xscale = "log",yscale = "log")
plt.plot(x,y,color = 'g',label = '$1/\sqrt{N}$')         #label每个plot指定一个字符串标签
plt.xlabel("N")
plt.legend(loc='best') 
ax=plt.gca()
ax.set_title('The deviation from $1/\sqrt{N}$',fontsize=14,color='r')
plt.scatter([10**2,10**3,10**4,10**5,10**6,10**7,10**8], [0.130890, 0.019491, 0.008378, 0.002419, 0.000540, 0.000131, 0.000092])
plt.scatter([10**2,10**3,10**4,10**5,10**6,10**7,10**8], [0.188710, 0.119408, 0.057815, 0.009767, 0.000194, 0.000579, 0.000184],color='r')
plt.show()
```
