```python
# -*- coding: utf-8 -*-
import numpy as np
import math
import matplotlib.pyplot as plt

plt.figure(num=2,dpi=300)

#
f1 = open('data.txt')
s1 = f1.read()
p1 = list(float(i) for i in s1.split())
x1 = p1[::2]
y1 = p1[1::2]


ax=plt.gca()
plt.plot(x1,y1,color='red',linewidth=1.0,linestyle='-')
#plt.xlim(-10,10)
#plt.ylim(0,1)
plt.xlabel('energy(eV)')
plt.ylabel('Intensity(counts)')
ax.set_title('Distribution',fontsize=14,color='r')
plt.show()
```

# "data.txt""direct_sampling"

```python
# -*- coding: utf-8 -*-
import numpy as np
import math
import matplotlib.pyplot as plt

plt.figure(num=2,dpi=300)

#
f1 = open('data.txt')
s1 = f1.read()
p1 = list(float(i) for i in s1.split())
x1 = p1[::2]
y1 = p1[1::2]

f2 = open('direct_sampling.txt')
s2 = f2.read()
p2 = list(float(i) for i in s2.split())
x2 = p2[::2]
y2 = list(100*i for i in p2[1::2])


ax=plt.gca()
plt.plot(x1,y1,color='red',linewidth=2.0,linestyle='-')
plt.plot(x2,y2,color='blue',linewidth=1.0,linestyle='-')

plt.xlabel('energy(eV)')
plt.ylabel('Intensity(counts)')
ax.set_title('Distribution',fontsize=14,color='r')
plt.show()
```
