```python
# -*- coding: utf-8 -*-
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
#plt.figure(num=2,dpi=300)

#画频数图
f = open('p1_2.txt')
s = f.read()
p = list(float(i) for i in s.split())

sigma = 0.447
N = 2
#画出理论图
x=np.arange(-1,1,0.05)
y=[  1/np.sqrt(2*np.pi*sigma/np.sqrt(N))*np.exp(-0.5*a**2/sigma**2*N)*180  for a in x]

#画出频数分布图
plt.hist(p, bins=200,range=(-1,1))


ax=plt.gca()
plt.plot(x,y,color='red',linewidth=1.0,linestyle='-')
plt.xlim(-1,1)
#plt.ylim(0,1)
plt.xlabel('x')
plt.ylabel('Distribution of random number')
ax.set_title('The first function distribution when N=%d, total=10000'%(N),fontsize=14,color='r')
plt.show()
```
