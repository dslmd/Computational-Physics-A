### 接下来是平均值后的相关函数

```python
p_phi = np.array(p_phi)
ux = np.cos(p_phi)
```

```python
def corr(t):
    total = 0
    for i in range(0,10000,1000):
        total += ux[t+i]*ux[0+i]
    total = total / 10
    return total
```

```python
x=range(0,1000,1)
y=[corr(a) for a in x]

ax=plt.figure(dpi=200)
plt.plot(x,y,color='b',linewidth=1.0,linestyle='-')
#plt.xlim(-10,10)
#plt.ylim(0,1)
plt.xlabel('t/ms')
plt.ylabel('Autocorrelation function')
plt.title('Autocorrelation function:C(t)=<$u_x(t)u_x(0)$>',fontsize=14,color='r')
plt.show()
```

### 验证theta防菌根正比于t
$$\sqrt{<\theta^2>}=Dt$$

```python
def theta(t):
    total = 0
    for i in range(0,10000,1000):
        total += p_theta[t+i]
    total = total / 10
    return total
```

```python
x=range(0,1000,1)
y=[theta(a) for a in x]
y2=[0.000075895*np.sqrt(a) for a in x]

ax=plt.figure(dpi=200)
plt.plot(x,y,color='b',linewidth=1.0,linestyle='-',label="$\sqrt{<\Theta^2>}$")
plt.plot(x,y2,color='r',label = '$0.000076\sqrt{t}$')
#plt.xlim(-10,10)
#plt.ylim(0,1)
plt.xlabel('t/ms')
plt.ylabel('$\sqrt{<\Theta^2>}$')
plt.legend()
plt.title('$\sqrt{<\Theta^2>}$',fontsize=14,color='r')
plt.show()
```
